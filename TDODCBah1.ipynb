Suppression des colonnes Web scrapper order et webscraper star url

```python
data.drop(['web-scraper-order', 'web-scraper-start-url'], axis=1, inplace=True)
data.head()
```

Suppression des symboles € de la colonne Prix

```python
data['Prix'] = data['Prix'].astype(str).str.replace('€', '', regex=False)
data
```

```python
data.dtypes
```

Convertir la colonne Prix en Float

```python
data['Prix'] = data['Prix'].str.replace(',', '.', regex=False).str.replace('\xa0', '', regex=False)
data['Prix'] = data['Prix'].astype(float)
data.dtypes
```
